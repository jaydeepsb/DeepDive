# create an example code and slides

````python
from numpy import arange
from pylab import plot, figure
from IPython.display import display, Math

x = arange(10)
y = x**2

display(Math('y = x^2'))
````

## Using nbagg as matplotlib backend will display interactive plots.

````python
%matplotlib nbagg
plot(x,y,"o-")
````

````python
figure()
plot(x,x, "o-")
````
